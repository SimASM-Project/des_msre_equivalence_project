# MSRE Verification: Fork-Join Topology

A **fork-join queueing network** with $n$ parallel branches. Arriving entities are split (forked) into $n$ sub-entities, each processed independently at a branch server. All sub-entities must complete before the entity is reassembled (joined). Each branch server has 5 servers.

## Macro-Step Refinement Equivalence (MSRE)

Each section below converts the JSON specification into SimASM via `%%simasm convert`, then verifies MSRE across all three formalism pairs.

**Notebook parameters**: `seed=42`, `run_length=100.0`.

```python
import simasm
assert simasm.__version__ >= "0.8.1", f"Requires simasm >= 0.8.1, got {simasm.__version__}"
```

## $n = 1$

```python
%%simasm convert
convert fork_join_1_eg:
    source: "../input/fork_join/fork_join_1_eg.json"
    formalism: event_graph
    register: "fork_join_1_eg"
    print: 30
endconvert
```

```python
%%simasm convert
convert fork_join_1_acd:
    source: "../input/fork_join/fork_join_1_acd.json"
    formalism: acd
    register: "fork_join_1_acd"
    print: 30
endconvert
```

```python
%%simasm convert
convert fork_join_1_devs:
    source: "../input/fork_join/fork_join_1_devs.json"
    formalism: devs
    register: "fork_join_1_devs"
    print: 30
endconvert
```

```python
%%simasm verify
// MSRE Verification: fork_join_1 EG vs ACD

verification fork_join_1_eg_acd_msre:

    models:
        import EG from "fork_join_1_eg"
        import ACD from "fork_join_1_acd"
    endmodels

    seed: 42

    labels:
        label queue_1_nonempty for EG: "queue_count_1 > 0"
        label queue_1_nonempty for ACD: "marking(Q_1) > 0"
        label server_1_busy for EG: "server_count_1 > 0"
        label server_1_busy for ACD: "marking(S_1) < num_servers"
    endlabels

    observables:
        observable queue_1_nonempty:
            EG -> queue_1_nonempty
            ACD -> queue_1_nonempty
        endobservable
        observable server_1_busy:
            EG -> server_1_busy
            ACD -> server_1_busy
        endobservable
    endobservables

    check:
        type: macro_step_refinement
        run_length: 100.0
        timeout: 60
    endcheck

    output:
        format: "json"
        include_counterexample: true
    endoutput

endverification
```
