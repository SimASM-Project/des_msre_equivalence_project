# MSRE Verification: Hybrid Topology

A **hybrid queueing network** combining $m$ tandem stations with a fork-join section of $b$ parallel branches (`hybrid_m_b`). This topology tests mixed serial-parallel structures. Each station has 5 servers.

## Macro-Step Refinement Equivalence (MSRE)

Each section below converts the JSON specification into SimASM via `%%simasm convert`, then verifies MSRE across all three formalism pairs.

**Notebook parameters**: `seed=42`, `run_length=100.0`.

```python
import simasm
assert simasm.__version__ >= "0.8.2", f"Requires simasm >= 0.8.2, got {simasm.__version__}"
```

```python
%%simasm verify
// MSRE Verification: hybrid_2_2 EG vs ACD

verification hybrid_2_2_eg_acd_msre:

    models:
        import EG from "hybrid_2_2_eg"
        import ACD from "hybrid_2_2_acd"
    endmodels

    seed: 42

    labels:
        label queue_1_nonempty for EG: "queue_count_1 > 0"
        label queue_1_nonempty for ACD: "marking(Q_1) > 0"
        label server_1_busy for EG: "server_count_1 > 0"
        label server_1_busy for ACD: "marking(S_1) < num_servers"
        label queue_2_nonempty for EG: "queue_count_2 > 0"
        label queue_2_nonempty for ACD: "marking(Q_2) > 0"
        label server_2_busy for EG: "server_count_2 > 0"
        label server_2_busy for ACD: "marking(S_2) < num_servers"
        label queue_3_nonempty for EG: "queue_count_3 > 0"
        label queue_3_nonempty for ACD: "marking(Q_3) > 0"
        label server_3_busy for EG: "server_count_3 > 0"
        label server_3_busy for ACD: "marking(S_3) < num_servers"
        label queue_4_nonempty for EG: "queue_count_4 > 0"
        label queue_4_nonempty for ACD: "marking(Q_4) > 0"
        label server_4_busy for EG: "server_count_4 > 0"
        label server_4_busy for ACD: "marking(S_4) < num_servers"
    endlabels

    observables:
        observable queue_1_nonempty:
            EG -> queue_1_nonempty
            ACD -> queue_1_nonempty
        endobservable
        observable server_1_busy:
            EG -> server_1_busy
            ACD -> server_1_busy
        endobservable
        observable queue_2_nonempty:
            EG -> queue_2_nonempty
            ACD -> queue_2_nonempty
        endobservable
        observable server_2_busy:
            EG -> server_2_busy
            ACD -> server_2_busy
        endobservable
        observable queue_3_nonempty:
            EG -> queue_3_nonempty
            ACD -> queue_3_nonempty
        endobservable
        observable server_3_busy:
            EG -> server_3_busy
            ACD -> server_3_busy
        endobservable
        observable queue_4_nonempty:
            EG -> queue_4_nonempty
            ACD -> queue_4_nonempty
        endobservable
        observable server_4_busy:
            EG -> server_4_busy
            ACD -> server_4_busy
        endobservable
    endobservables

    check:
        type: macro_step_refinement
        run_length: 100.0
        timeout: 60
    endcheck

    output:
        format: "json"
        include_counterexample: true
    endoutput

endverification
```
