# MSRE Verification: Feedback Topology

A **feedback queueing network** with $n$ stations in series. After completing service at station $n$, entities are routed back to station 1 with a fixed probability, creating recirculation. Each station has 5 servers (`service_capacity=5`).

## Macro-Step Refinement Equivalence (MSRE)

Each cell below verifies that two formalism translations produce **identical observable labels at every tick boundary**.

**Notebook parameters**: `seed=42`, `run_length=100.0`.

```python
import simasm
assert simasm.__version__ >= "0.7.0", f"Requires simasm >= 0.7.0, got {simasm.__version__}"
```

## $n = 1$

```python
%%simasm verify
// MSRE Verification: feedback_1 EG vs ACD
// Generated by generate_msre_configs.py

verification feedback_1_eg_acd_msre:

    models:
        import EG from "../models/eg/feedback/feedback_1_eg.simasm"
        import ACD from "../models/acd/feedback/feedback_1_acd.simasm"
    endmodels

    seed: 42

    labels:
        label queue_1_nonempty for EG: "queue_count_1 > 0"
        label queue_1_nonempty for ACD: "marking(Q_1) > 0"
        label server_1_busy for EG: "server_count_1 > 0"
        label server_1_busy for ACD: "marking(S_1) < num_servers"
    endlabels

    observables:
        observable queue_1_nonempty:
            EG -> queue_1_nonempty
            ACD -> queue_1_nonempty
        endobservable
        observable server_1_busy:
            EG -> server_1_busy
            ACD -> server_1_busy
        endobservable
    endobservables

    check:
        type: macro_step_refinement
        run_length: 100.0
        timeout: 60
    endcheck

    output:
        format: "json"
        include_counterexample: true
    endoutput

endverification
```
